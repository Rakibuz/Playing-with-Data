```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv("avrg_custom.csv")

# Create the figure and the two axes
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# Plot the data on the two axes
ax1.plot(df['Month'], df['Year'], 'g-')
ax2.plot(df['Month'], df['avg_tavg'], 'b-')

# Set the labels for the x and y axes
ax1.set_xlabel('Month')
ax1.set_ylabel('Year', color='g')
ax2.set_ylabel('avg_tavg', color='b')

# Set the title for the plot
plt.title('Title of the plot')

# Show the plot
plt.show()
```
